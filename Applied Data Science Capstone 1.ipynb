For this part of the assignment it is required to scrap data from the Wikipedia page containing the postal codes of Canada. For this I will use `read_html` from the pandas library.

```python
#Load Pandas
import pandas as pd

#Creation of the url containing the data:
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

#Reading the url and storing it on a list called data_raw:
data_raw = pd.read_html(url)
data_raw
```

The result is a list that contains several dataframes. The index that seems to contain the dataframe needed to work is on the **index 0** of the list:

```python
data_raw[0]
```

Having figured out the data needed to start, I save it on a new dataframe to process it untill it's ready to be worked with:

```python
df = data_raw[0]
df.head()
```

Now the dataframe is modified by filtering out all the Borough rows that are "Not assigned." Then the index is reset, dropping the old index and finally the first 12 rows are shown.

```python
df = df[df["Borough"] != "Not assigned"].reset_index(drop = True)
df.head(12)
```
