I will be using the Toronto Neighbourhood profiles from 2016 obtained from Toronto's official webpage.

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
df = pd.read_csv('neighbourhood-profiles-2016-csv.csv')
```

First, some initial cleanup:

```python
#Replace all of , for empty:
df = df.replace(',', '', regex = True)
df
```

# Feature selection
For the analysis I want to make, I want to use the population of each neighbourhood, their income and the distribution of ages. I have the conjecture these factors might affect the type of restaurant that's favored on each neighbourhood. The income might affect the price of the preffered restaurants. The age distributions, the type of restaurant that's more popular in each neighbourhood.

This is the official 2016 census, it has a lot of information. It has several columns that allow me to know beforehand what sort of information is collected. For instance, the Category column has this information:

```python
df.Category.value_counts()
```

For this particular analysis and for the sake of having a manageable dataframe for my current skill level. I will not account factors such as Language, Ethics or any race related metrics. This is not a politically motivated reason, it's simply a practical one, given the afromentioned categories take a good amount of space in the dataframe. 

In future projects when I have more knowledge and can use more robust machine learning tools; along with more programming expertise, I would like take into account the categories I'm ignoring on this project.

```python
#Dropping categories to scale down the data:
df = df[(df.Category != 'Language') &
        (df.Category != 'Language of work') &
        (df.Category != 'Ethnic origin') &
        (df.Category != 'Immigration and citizenship') &
        (df.Category != 'Aboriginal peoples') &
        (df.Category != 'Visible minority')
       ]
```

The remaining categories are:

```python
df.Category.unique()
```

There are still some categories left that might give interesting insights. However, I will focus only in the income and population:

```python
income_df = df[df.Category == 'Income']
```

```python
population_df = df[df.Category == 'Population']
```

Further exploration of the population dataframe:

```python
population_df.Topic.value_counts()
```

```python
population_df.Characteristic.unique()
```

I'm interested in the Population of 2016, Population density, and the age distributions of the population:

```python
population_df = population_df.loc[[2, 7]].append(population_df.loc[9:14])
```

```python
population_df
```

I have what I want from the population statistics. For now I set it aside and process the income dataframe.

```python
income_df.Topic.value_counts()
```

The information is very granular, once again lots of information to make plenty of analysis. After perusing the categories, I decide to explore the Total income and After-tax income:

```python
income_df = income_df[income_df.Topic == 'Income sources']
```

```python
income_df[(income_df.Characteristic == 'Total income: Average amount ($)')].append(income_df[(income_df.Characteristic == '  After-tax income: Average amount ($)')])
```

I decide to go with After-tax income.

```python
income_df = income_df[(income_df.Characteristic == '  After-tax income: Average amount ($)')]
```

Append the income and population dataframe, with the neighbourhood number:

```python
toronto_merged = population_df.append(income_df)
```

```python
#Grab the neighbourhood number row.
num = df.loc[0].to_frame().transpose()
num
```

```python
toronto_merged = num.append(toronto_merged)
```

```python
#Save to a csv file the current state of the dataframe in case something happens.
toronto_merged.to_csv('toronto_merged_raw.csv', index = False)
toronto_merged
```

Time to clean up the dataframe, make sure the columns have the correct data type and drop columns that are no longer needed.

```python
#Drop columns
columns_to_drop = ['_id', 'Category', 'Topic', 'Data Source', 'City of Toronto']
toronto_merged.drop(columns_to_drop, axis = 1, inplace = True)
```

```python
#Transpose the dataframe
toronto_merged = toronto_merged.transpose()
```

```python
#Index and column cleanup after the transpose
toronto_merged.reset_index(inplace = True)
toronto_merged.columns = toronto_merged.loc[0]
toronto_merged.drop(0, axis = 0, inplace = True)
toronto_merged.reset_index(drop = True, inplace = True)
toronto_merged.rename({'Characteristic': 'Neighbourhood'}, axis = 1, inplace = True)
```

```python
#Strip trailing and leading spaces on the columns:
toronto_merged.columns = toronto_merged.columns.str.strip()
```

```python
#Pop the first column to typecast all of the dataframe at once.
temp_col = toronto_merged.pop('Neighbourhood')
toronto_merged.head()
```

```python
#Typecast the frame, then return the first column
toronto_merged = toronto_merged.astype(int)
toronto_merged.insert(0, 'Neighbourhood', temp_col)
```
