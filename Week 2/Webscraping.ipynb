# Web Scrapping
Fetching all the libraries I need to scrap the webpage:

```python
#Libraries for scrapping 
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium import webdriver
import selenium
import pandas as pd
from bs4 import BeautifulSoup

#Libraries to assign neighbourhoods
import json
from shapely.geometry import shape, Point
```

```python
#Setting up the webdriver
options = webdriver.FirefoxOptions()
options.set_headless(True)
driver = webdriver.Firefox(options = options)
```

```python
#Defining the base url
url = "https://www.point2homes.com/CA/Commercial-Properties-For-Lease.html?location=Toronto,%20ON&PropertyType=Commercial&search_mode=location&Keywords=Retail&SelectedView=listings&LocationGeoId=783094&viewport=-79.84037484208983,43.519311936783744,-78.98962105791014,43.92176999910959&zoom=11&page="
```

```python
#Defining a function to add the page number at the end of the url.
def add_url(page):
    return url + str(page)
```

```python
#Define the empty dictionary to save the results:
temp = []

#Scrape 12 pages
for page in range(1,13):
    #Access the webpage:
    driver.get(add_url(page))
    
    #Get the amount of results per page:
    list_cnts = driver.find_elements_by_xpath("//div[@class = 'item-cnt clearfix  ']")
    
    #Go over each result of the webpage:
    for i in range(len(list_cnts)):
        address = driver.find_elements_by_xpath("//div[@class = 'item-right-details-cnt']/div[1]")[i].text
        area = driver.find_elements_by_xpath("//div[@class = 'item-right-details-cnt']/div[2]")[i].text
        rent = driver.find_elements_by_xpath("//div[@class = 'item-right-utils-cnt']/*/span[@class = 'green']")[i].text
        
        #While inspecting the page, I noticed there was a hidden element that was storing the coordinates of
        #the location. The only way I found out to get those out was by getting the raw html of that particular element,
        #then use BeautifulSoup to parse the HTML, then get the coordinates:
        
        #Getting the html of the element:
        right_html = driver.find_elements_by_xpath("//div[@class = 'item-right-utils-cnt']")[i].get_attribute('innerHTML')
        
        #Digging out the coordinates:
        coor_html = BeautifulSoup(right_html)
        try:
            lat = coor_html.find_all('input')[1].attrs['value']
            lon = coor_html.find_all('input')[2].attrs['value']
        except:
            lat = 'None'
            lon = 'None'
        
        #Storing the results on a dictionary:
        temp.append({"Address": address, "Area": area, 'Rent': rent, 'Latitude': lat, 'Longitude': lon})

#Convert to a dataframe the dictionary.        
rentals = pd.DataFrame(temp)
driver.close()
print('Done')
```

```python
#Geocoding
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='Rental_venues_locations')
from geopy.extra.rate_limiter import RateLimiter
geocoder = RateLimiter(geolocator.geocode, min_delay_seconds=1.5)
```

```python
#Save the addresses with missing coordinates to a new frame to modify it:
df_m = df[df.Latitude == 'None']
df_m.reset_index(inplace = True, drop = True)
```

```python
#Drop the rows with no coordinates from the compiled dataframe:
df = df[df.Latitude != 'None']
```

```python
#When Geopy uses Nominatin, it usually fails to return a coordinate when the address is wrong, or is pointing 
#to something specific like a particular office, warehouse or floor. I check wich address have that sort info:
def check_address():
    for i in range(df_m.shape[0]):
        try:
            df_m.loc[i, 'Latitude'] = geocoder(df_m.Address[i]).latitude
            df_m.loc[i, 'Longitude'] = geocoder(df_m.Address[i]).longitude
            print(i, 'OK')
        except:
            print(i, "Wrong address")
```

# Adding neighbourhood number
For this project I refrain from using a geocoder to get the neighbourhood names, because I've taken the GEOjsons and the Toronto census from the official webpage, in those sources the neighbourhoods are coded in the same way. Using a geocoder would mean I wouldn't get the same neighbourhood name or code; it's something I tried and decided not to purse any further.

The purpose of wrangling this data is to get the average rental value per square feet of current, available commercial locations in Toronto. I decided to scrap two webpages instead of using an API, simply because I wanted to learn and practice web scrapping; also, most real states API have fees. This aproach may yield a smaller amount of data, but for the sake of cutting my teeth in as many aspects of a Data Science project as possible; scrapping the data my self gave me greater insight into the work that goes into a project.

To get the neighbourhood based only on the location data, I stumbled into a code that determines whether or not a coordinate lies inside the polygon defined by the GEOjson geometry data. Using that algoritm, I can use it to assign the Neighbourhood number according to the official Toronto codification.

```python
#load GeoJSON file containing sectors:
with open('Toronto_Neighbourhoods.geojson') as f:
    js = json.load(f)
```

```python
#Add the empty Neighbourhood column to the dataframe:
df_clean['Neighbourhood Number'] = 0
```

```python
#Iterate on each row of the dataframe:
for i in range(len(df_clean.Address)):
    #Construct point based on lon/lat returned by geocoder:
    point = Point(df_clean.Longitude[i], df_clean.Latitude[i])
    
    #Check each polygon to see if it contains the point
    for feature in js['features']:
        polygon = shape(feature['geometry'])
        if polygon.contains(point):
            df_clean.loc[i, 'Neighbourhood Number'] = feature['properties']['AREA_SHORT_CODE']
```
